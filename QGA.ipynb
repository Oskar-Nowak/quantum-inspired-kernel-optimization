```python
from QGA import QGA
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = load_wine()
X = StandardScaler().fit_transform(data.data)
X_train, _, y_train, _ = train_test_split(X, data.target, test_size=0.3, random_state=42)

qga = QGA(X_train, y_train, generations=300, population_size=20, genome_length=24)
qga.run()
qga.plot_output()
```
